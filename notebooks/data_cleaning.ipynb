```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named df
# Group by 'primary_category' and count the number of products
category_counts = df['primary_category'].value_counts()

# Plot the data using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=category_counts.index, y=category_counts.values, palette="viridis")
plt.title('Number of Products per Primary Category')
plt.xlabel('Primary Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

# Breakdown of Secondary Categories for Each Primary Category

```python

# Drop rows where 'secondary_category' is NaN
df_filtered = df.dropna(subset=['secondary_category'])

# Group by 'primary_category' and 'secondary_category' and count the number of products
category_counts = df_filtered.groupby(['primary_category', 'secondary_category']).size().reset_index(name='count')

# Get a list of unique primary categories
primary_categories = category_counts['primary_category'].unique()

# Define the number of rows and columns for the grid layout
num_rows = len(primary_categories)
num_cols = 1  # One column for each primary category

# Increase the figure size
plt.figure(figsize=(15, 5 * num_rows))

for i, primary_category in enumerate(primary_categories, 1):
    plt.subplot(num_rows, num_cols, i)
    # Filter data for the current primary category
    data = category_counts[category_counts['primary_category'] == primary_category]
    sns.barplot(x='secondary_category', y='count', data=data, palette='viridis')
    plt.title(f'Number of Products per Secondary Category in {primary_category}')
    plt.xlabel('Secondary Category')
    plt.ylabel('Number of Products')
    plt.xticks(rotation=45, ha='right')

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```

```python
import ast

# Sample skin types

skin_types = ['dry', 'combo', 'normal', 'oily']

# Function to check if a skin type is in the highlights
def check_skin_type(highlights, skin_type):
    if pd.isna(highlights):  # Handle NaN values
        return 0
    try:
        highlights_list = ast.literal_eval(highlights.lower())  # Convert string representation of list to an actual list and lowercase it
        for item in highlights_list:
            if skin_type in item:
                return 1
    except (ValueError, SyntaxError):  # Catch errors if highlights is not a valid list string
        return 0
    return 0
```

```python
# Function to check if "Vegan" is in the highlights
def check_vegan(highlights):
    if pd.isna(highlights):  # Handle NaN values
        return 0
    try:
        highlights_list = ast.literal_eval(highlights.lower())  # Convert string representation of list to an actual list and lowercase it
        if 'vegan' in highlights_list:
            return 1
    except (ValueError, SyntaxError):  # Catch errors if highlights is not a valid list string
        return 0
    return 0
```

```python
# Create a new column for each skin type and populate it
for skin_type in skin_types:
    column_name = skin_type
    df[column_name] = df["highlights"].apply(lambda x: check_skin_type(x, skin_type))

# Add the Vegan column
df['vegan'] = df['highlights'].apply(check_vegan)
```
